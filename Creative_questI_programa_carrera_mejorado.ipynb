```python
# ==========================================
# Programa: Cobertura con Forward cambiario
# Autor: Estudiante de Ingeniería Financiera
# Objetivo: Calcular la diferencia entre recibir dólares con y sin cobertura
# Versión con interfaz gráfica (tkinter)
# ==========================================

import tkinter as tk
from tkinter import messagebox

def cobertura_forward(monto_usd, tasa_forward, tasas_mercado):
    """
    Calcula el valor recibido por la empresa con y sin cobertura.
    """
    resultados = {}

    # Valor asegurado con el forward (siempre fijo)
    valor_forward = monto_usd * tasa_forward
    resultados["Cobertura_Forward"] = valor_forward

    # Valores sin cobertura (según mercado)
    valores_sin_cobertura = [monto_usd * t for t in tasas_mercado]
    resultados["Sin_Cobertura"] = valores_sin_cobertura

    return resultados


# ==========================================
# Función para usar en la interfaz
# ==========================================
def ejecutar_calculo():
    try:
        monto_usd = float(entry_monto.get())
        tasa_forward = float(entry_forward.get())

        # Obtener tasas separadas por coma
        tasas_mercado = list(map(float, entry_tasas.get().split(",")))

        # Cálculo usando la función original
        resultados = cobertura_forward(monto_usd, tasa_forward, tasas_mercado)

        # Mostrar resultados
        salida = f"=== RESULTADOS COBERTURA ===\n"
        salida += f"Monto en USD: {monto_usd}\n"
        salida += f"Tasa Forward pactada: {tasa_forward} COP/USD\n\n"

        salida += "--- Sin cobertura ---\n"
        for tasa, valor in zip(tasas_mercado, resultados["Sin_Cobertura"]):
            salida += f"Tasa mercado {tasa} → Valor recibido: {valor:,.0f} COP\n"

        salida += "\n--- Con cobertura (Forward) ---\n"
        salida += f"Valor recibido asegurado: {resultados['Cobertura_Forward']:,.0f} COP"

        # Mostrar en el cuadro de texto
        text_resultado.delete("1.0", tk.END)
        text_resultado.insert(tk.END, salida)

    except ValueError:
        messagebox.showerror("Error", "Por favor ingrese números válidos.\nLas tasas deben separarse por comas.")


# ==========================================
# Interfaz gráfica (tkinter)
# ==========================================
ventana = tk.Tk()
ventana.title("Cobertura con Forward Cambiario")
ventana.geometry("520x500")
ventana.configure(bg="#e6f0ff")

# Título principal
tk.Label(ventana, text="Cobertura con Forward Cambiario",
         font=("Arial", 14, "bold"), bg="#e6f0ff", fg="#003366").pack(pady=10)

# Frame para entradas
frame = tk.Frame(ventana, bg="#e6f0ff")
frame.pack(pady=5)

tk.Label(frame, text="Monto en USD:", bg="#e6f0ff").grid(row=0, column=0, sticky="e", pady=5)
entry_monto = tk.Entry(frame, width=25)
entry_monto.grid(row=0, column=1)

tk.Label(frame, text="Tasa Forward (COP/USD):", bg="#e6f0ff").grid(row=1, column=0, sticky="e", pady=5)
entry_forward = tk.Entry(frame, width=25)
entry_forward.grid(row=1, column=1)

tk.Label(frame, text="Tasas de mercado (separadas por comas):", bg="#e6f0ff").grid(row=2, column=0, sticky="e", pady=5)
entry_tasas = tk.Entry(frame, width=25)
entry_tasas.grid(row=2, column=1)

# Botón para ejecutar cálculo
tk.Button(ventana, text="Calcular", bg="#3366cc", fg="white",
          font=("Arial", 10, "bold"), command=ejecutar_calculo).pack(pady=10)

# Cuadro de texto para resultados
text_resultado = tk.Text(ventana, height=15, width=60, bg="white", fg="black")
text_resultado.pack(pady=10)

# Iniciar la ventana
ventana.mainloop()
```

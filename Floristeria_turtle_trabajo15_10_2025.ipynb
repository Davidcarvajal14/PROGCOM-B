```python
import turtle

# Configuracin inicial
screen = turtle.Screen()
screen.title("Floristera con jardn interior por zonas")
screen.bgcolor("lightyellow")
t = turtle.Turtle()
t.speed(0)

# Funciones bsicas
def dibujar_rectangulo(t, ancho, alto, color):
    t.fillcolor(color)
    t.begin_fill()
    for _ in range(2):
        t.forward(ancho)
        t.left(90)
        t.forward(alto)
        t.left(90)
    t.end_fill()

def dibujar_triangulo(t, base, altura, color):
    t.fillcolor(color)
    t.begin_fill()
    t.forward(base)
    t.left(135)
    t.forward(altura)
    t.left(90)
    t.forward(altura)
    t.left(135)
    t.end_fill()

def dibujar_maceta_con_flor(t, x, y, flor_color):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.color("saddlebrown")
    t.pendown()
    dibujar_rectangulo(t, 10, 5, "saddlebrown")
    t.penup()
    t.goto(x + 5, y + 5)
    t.setheading(0)
    t.color(flor_color)
    t.pendown()
    for _ in range(6):
        t.begin_fill()
        t.circle(3)
        t.end_fill()
        t.left(60)

def dibujar_arbol(t, x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.color("saddlebrown")
    t.pendown()
    dibujar_rectangulo(t, 5, 20, "saddlebrown")
    t.penup()
    t.goto(x - 10, y + 20)
    t.color("forestgreen")
    t.begin_fill()
    t.circle(15)
    t.end_fill()

def dibujar_techo_simple(t, x, y, base, altura, color):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, base, altura, color)

# Vista frontal
def vista_frontal(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "mistyrose")
    dibujar_techo_simple(t, x, y + 100, 100, 70, "saddlebrown")

    # Puerta
    t.penup()
    t.goto(x + 35, y)
    t.pendown()
    dibujar_rectangulo(t, 30, 50, "peru")

    # Ventana central
    t.penup()
    t.goto(x + 40, y + 60)
    t.pendown()
    dibujar_rectangulo(t, 20, 20, "white")

    # Letrero "Floristera"
    t.penup()
    t.goto(x + 30, y + 85)
    t.color("black")
    t.write("Floristera", font=("Arial", 8, "bold"))

    # Flores dentro de la casa, a los lados de la puerta
    posiciones_flores = [x + 5, x + 20, x + 75, x + 85]
    colores = ["red", "blue", "purple", "orange"]
    for px, color in zip(posiciones_flores, colores):
        dibujar_maceta_con_flor(t, px, y + 10, color)

# Vista trasera con rboles dentro
def vista_trasera(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "mistyrose")
    dibujar_techo_simple(t, x, y + 100, 100, 70, "saddlebrown")

    # rboles dentro de la casa
    posiciones_arboles = [x + 10, x + 30, x + 50, x + 70, x + 90]
    for px in posiciones_arboles:
        dibujar_arbol(t, px, y + 5)

# Vista lateral izquierda con flores dentro
def vista_lateral_izquierda(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "mistyrose")
    dibujar_techo_simple(t, x, y + 100, 60, 42, "saddlebrown")

    # Flores dentro de la casa
    posiciones = [x + 5, x + 20, x + 35, x + 50]
    colores = ["blue", "magenta", "yellow", "deeppink"]
    for px, color in zip(posiciones, colores):
        dibujar_maceta_con_flor(t, px, y + 10, color)

# Vista lateral derecha con flores dentro
def vista_lateral_derecha(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "mistyrose")
    dibujar_techo_simple(t, x, y + 100, 60, 42, "saddlebrown")

    # Flores dentro de la casa
    posiciones = [x + 5, x + 20, x + 35, x + 50]
    colores = ["purple", "orange", "limegreen", "hotpink"]
    for px, color in zip(posiciones, colores):
        dibujar_maceta_con_flor(t, px, y + 10, color)

# Vista superior
def vista_superior(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 80, "saddlebrown")

# Dibujo completo
vista_frontal(-250, 100)
vista_trasera(-80, 100)
vista_lateral_izquierda(100, 100)
vista_lateral_derecha(220, 100)
vista_superior(-50, -150)

t.hideturtle()
screen.mainloop()
```
